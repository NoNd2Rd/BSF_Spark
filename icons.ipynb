## 🧪 Example Python Usage

```python
# Define emojis for logging
EMOJI_LOGGING = {
    "start": "🚀",
    "in_progress": "⏳",
    "complete": "✅",
    "success": "🎉",
    "error": "❌",
    "warning": "⚠️",
    "info": "ℹ️",
    "question": "❓",
}

# Example function
def run_process():
    print(f"{EMOJI_LOGGING['start']} Starting process...")
    print(f"{EMOJI_LOGGING['in_progress']} Process is running...")
    print(f"{EMOJI_LOGGING['complete']} Process completed successfully!")
    print(f"{EMOJI_LOGGING['success']} All tasks succeeded!")

run_process()
